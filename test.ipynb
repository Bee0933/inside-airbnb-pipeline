```python
# calendar
dt = merge_data(listings_csv_dir).head()
```

```python
df = dt.copy()
```

```python
df.last_scraped = pd.to_datetime(df.last_scraped)
df.description = df.description.str.replace(r'<[^<>]*>', f'{os.linesep}', regex=True)
df.host_since = pd.to_datetime(df.host_since)
df.host_about = df.host_about.str.replace(r'\n', f'', regex=True).replace(r'\r', f'', regex=True)
df.host_response_rate = df.host_response_rate.str.replace('%', '', regex=True).astype('float16')
df.host_acceptance_rate = df.host_acceptance_rate.str.replace('%', '', regex=True).astype('float16')
df["host_verifications_phone"] = np.where("phone" in str(df.host_verifications), 't', 'f').astype('object')
df["host_verifications_email"] = np.where("email" in str(df.host_verifications), 't', 'f').astype('object')
df = df.drop(['host_verifications'], axis=1)
df.bedrooms = df.bedrooms.astype('int64')
df.beds = df.beds.astype('int64')
df.price = df.price.str.replace('$', '', regex=True).astype('float64')

# amenities
amenit=list(df.amenities.unique())
temp_1 = ast.literal_eval(amenit[0])
for i in amenit[1:]:
    temp = ast.literal_eval(i)
    result = list(set(temp) | set(temp_1))
    temp_1 = result

col_result = [val.replace(' ', '_').replace('-', '_').replace('â€“','') for val in result]
for column_head, column in zip(col_result, result):
    df[f"{column_head}"] = np.where(column in str(df.amenities), 't', 'f').astype('object')
df = df.drop(['amenities'], axis=1)

df.minimum_minimum_nights = df.minimum_minimum_nights.astype('int64')
df.maximum_minimum_nights = df.maximum_minimum_nights.astype('int64')
df.minimum_maximum_nights = df.minimum_maximum_nights.astype('int64')
df.maximum_maximum_nights = df.maximum_maximum_nights.astype('int64')


df = df.drop(['calendar_last_scraped'], axis=1)
df.first_review = pd.to_datetime(df.first_review)
df.last_review = pd.to_datetime(df.last_review)


df = df.rename(columns={'host_response_rate': 'host_response_rate_%', 'host_acceptance_rate': 'host_acceptance_rate_%'})
```

## remove old data

```python
rm_path = "data/processed/listings"
ext = "parquet"

def delete_files_and_dir(dir:Path, file_extension:str)->List[Path]:
    files_in_dir = []
    for f in glob.glob(f"{rm_path}/*.{file_extension}"):
        os.remove(f)
        files_in_dir.append(f)
    os.removedirs(rm_path)
    return f" deleted files : {files_in_dir}"
```
